# Data Preparation 

```python
#load csv
import pandas as pd
flights = pd.read_csv('selected_flights.csv')
```

```python
flights.head()
```

```python

selected_flights = flights.query("from_airport_code in ['PEK', 'CDG', 'YYZ', 'GRU', 'SYD', 'CAI'] or dest_airport_code in ['PEK', 'CDG', 'YYZ', 'GRU', 'SYD', 'CAI']")
```

```python
selected_flights.tail(5)
```

```python
selected_flights.to_csv('selected_flights.csv', index=False)
```

```python
#Morning 5am to 12pm 
#Afternoon 12pm to 5pm
#Evening 5pm to 10pm
#Night 10pm to 5am

def categorize_time(time):
    hour = pd.to_datetime(time).hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'
    
selected_flights['departure_time_category'] = selected_flights['departure_time'].apply(categorize_time)
selected_flights['arrival_time_category'] = selected_flights['arrival_time'].apply(categorize_time)

print(selected_flights)
```

```python
# Columns to one-hot encode
columns_to_encode = ["from_airport_code", "dest_airport_code", "departure_time_category", "arrival_time_category"]

# Perform one-hot encoding
selected_flights = pd.get_dummies(selected_flights, columns=columns_to_encode, dtype=int)

# Show the first few rows of the encoded DataFrame
print(selected_flights.head(5))
```

```python
from collections import defaultdict
import numpy as np

# Initialize a defaultdict to store the counts of each aircraft type
aircraft_counts = defaultdict(int)

# Iterate over the list of aircraft types in each row
for aircraft_list in selected_flights['aircraft_type'].str.split('|'):
    if isinstance(aircraft_list, list):  # Check if the value is a list
        # Increment the count for each aircraft type
        for aircraft_type in aircraft_list:
            aircraft_counts[aircraft_type] += 1

# Print the counts of each aircraft type
for aircraft_type, count in aircraft_counts.items():
    print(f"{aircraft_type}: {count}")

len(aircraft_counts)
```

```python
import pandas as pd

# Replace NaN values with an empty string
selected_flights['aircraft_type'] = selected_flights['aircraft_type'].fillna('')

# Create a set of unique aircraft types
unique_aircraft_types = set()
for aircraft_list in selected_flights['aircraft_type'].str.split('|'):
    if isinstance(aircraft_list, list):
        unique_aircraft_types.update(aircraft_list)

# Create new columns for each aircraft type
for aircraft_type in unique_aircraft_types:
    selected_flights[aircraft_type] = selected_flights['aircraft_type'].str.contains(aircraft_type).astype(int)

# Drop the original 'aircraft_type' column
selected_flights.drop(columns=['aircraft_type'], inplace=True)

# Print the updated DataFrame
print(selected_flights)
```

```python
from collections import defaultdict

# Initialize a defaultdict to store the counts of each airline
airline_counts = defaultdict(int)

# Iterate over the list of airline names in each row
for airline_list in selected_flights['airline_name'].str.split('|'):
    if isinstance(airline_list, list):  # Check if the value is a list
        # Remove square brackets and increment the count for each airline name
        for airline_name in airline_list:
            airline_name = airline_name.strip("[]")  # Strip square brackets
            airline_counts[airline_name] += 1

# Print the counts of each airline
for airline_name, count in airline_counts.items():
    print(f"{airline_name}: {count}")

# Get the number of unique airlines
num_unique_airlines = len(airline_counts)
print("Number of unique airlines:", num_unique_airlines)


```

```python
popular_airlines_list = [airline_name for airline_name, count in airline_counts.items() if count >= 2000]
len(popular_airlines_list)
```

```python
all_airlines_list = [airline_name for airline_name in airline_counts.keys()]
print(all_airlines_list)
```

```python
import pandas as pd
import re

# Replace NaN values with an empty string
selected_flights['airline_name'] = selected_flights['airline_name'].fillna('')

# Remove extra characters like '[', ']', and spaces from airline names
selected_flights['airline_name'] = selected_flights['airline_name'].str.replace('[\[\] ]', '')

# Create a set of unique airline names
unique_airline_names = set()
for airline_list in selected_flights['airline_name'].str.split('|'):
    if isinstance(airline_list, list):
        unique_airline_names.update(airline_list)

# Create new columns for each airline name
for airline_name in unique_airline_names:
    # Escape special characters in the airline name
    airline_name_escaped = re.escape(airline_name)
    selected_flights[airline_name] = selected_flights['airline_name'].str.contains(airline_name_escaped).astype(int)

# Drop the original 'airline_name' column
selected_flights.drop(columns=['airline_name'], inplace=True)

# Print the updated DataFrame
print(selected_flights)
# saving it into a csv to check the one-hot-encoding worked as planned
selected_flights.to_csv('selected_flights_one_hot_encoded.csv', index=False)
```
